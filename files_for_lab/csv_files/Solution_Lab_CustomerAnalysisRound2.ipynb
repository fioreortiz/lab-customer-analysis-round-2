```python
#6. Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.
marketing['months'] = marketing['effective_to_date']
marketing['months'] = pd.to_datetime(marketing['effective_to_date'], errors='coerce')
display(marketing.head())
```

```python
first_quarter = marketing[(marketing['months'].dt.month >= 1) & (marketing['months'].dt.month <= 3)]
display(first_quarter)
print(first_quarter.shape)
```

```python
# The shape of the dataframe is the same after the filter, and is because the data are just from january and february. 
display(marketing['months'].value_counts(dropna=False)) 
len(marketing[marketing['months'].isna()==True]) 
```
